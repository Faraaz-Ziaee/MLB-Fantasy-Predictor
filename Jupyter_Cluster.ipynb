```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import Normalizer

# Read in csv File to dataframe
df = pd.read_csv('Batting_Cleaned_Last_Year_Cut.csv')

# Only use 2018 for clustering analysis
df = df[df['yearID'] == 2018]

# Drop features that are not going to be used in clustering
df = df.drop(['yearID', 'AB', 'H', '2B', '3B', 'BB', 'SO', 'IBB', 'HBP', 'SF', 'OPS+', 'Result'], axis=1)
# Set index as playerID to retain info once clustering anaysis is complete
df.set_index(["playerID"], inplace = True, drop = True)

# Drop features not required for this specific analysis
x = df.drop(['R', 'RBI', 'SLG', 'OPS'], axis = 1)

# Show elbow method to determine how many clusters to use
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Run k-means clustering algorithm with 5 cluster centers
kmeans = KMeans(n_clusters=4, init='k-means++', n_init=10, random_state=0)
pred_y = kmeans.fit_predict(x)
plt.scatter(x.iloc[:,0], x.iloc[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],  c='red')
plt.title('Home Runs vs. OBP Cluster')
plt.xlabel('Home Runs')
plt.ylabel('OBP')
plt.show()
```

```python
# Run k-means clustering algorithm with 5 cluster centers
kmeans = KMeans(n_clusters=5, init='k-means++', n_init=10, random_state=0)
pred_y = kmeans.fit_predict(x)
plt.scatter(x.iloc[:,0], x.iloc[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],  c='red')
plt.title('Home Runs vs. OBP Cluster')
plt.xlabel('Home Runs')
plt.ylabel('OBP')
plt.show()
```
